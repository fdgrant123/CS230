```python
def createAvail(ppg, _min, _3pa, _2pa, paint, ft, games, players):
    for player in players:
        for game in games:
            if player in active:
                player_games = pd.read_csv(str(int(player)) + ".csv")
                game_i = player_games.index[player_games["Game_ID"] == game].tolist()
                if len(game_i) > 0:
                    game_id = game_i[0]
                    ppg += player_games["PTS"].iloc[game_id]
                    _min += float(player_games["MIN"].iloc[game_id].replace(":", "."))
                    _3pa += player_games["FG3A"].iloc[game_id]
                    _2pa += (player_games["FGA"].iloc[game_id] - player_games["FG3A"].iloc[game_id])
                    paint += player_games["PTS_PAINT"].iloc[game_id]
                    ft += player_games["FTM"].iloc[game_id]
                    
    return ppg, _min, _3pa, _2pa, paint, ft
    
```
